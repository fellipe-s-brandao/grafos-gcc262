```python
# 🧩 1. Importar bibliotecas
import pandas as pd
```

```python
# 📥 Ler o CSV
df = pd.read_csv('src/resultados_metricas_grafos.csv')

from IPython.display import display

# 👇 Isso exibe a tabela formatada bonitinha no notebook
display(df)
```

```python
# 📥 Ler o CSV de intermediação
df_intermediacao = pd.read_csv('src/resultados_intermediacao.csv')

# 🔍 Obter todos os nomes únicos de grafos
nomes_grafos = df_intermediacao['NomeGrafo'].unique()

# 📊 Separar e exibir os dados de cada grafo
for nome_grafo in nomes_grafos:
    print(f"📁 Grafo: {nome_grafo}")
    
    df_grafo = df_intermediacao[df_intermediacao['NomeGrafo'] == nome_grafo]
    
    # Exibe como tabela formatada
    display(df_grafo.reset_index(drop=True))  # remove o índice original e reorganiza a visualização
```
